# Raw data understanding
This jupyter notebook analyses and describes the raw dataset of the computer vision task for object detection of bone fractures of the FracAtlas dataset.


```python
import os
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import json
```

```python
dataset_path = "./raw data"
```

The raw data is already clustered in train, test and validation sets. In addition there is are non-fractured bone images.
For each set there are two subfolders with:

a) the images  
b) the annotations


```python
# Show subfolder structure
for root, dirs, files in os.walk(dataset_path):
    level = root.replace(dataset_path, '').count(os.sep)
    indent = ' ' * 4 * level
    print(f"{indent}{os.path.basename(root)}/")
    sub_indent = ' ' * 4 * (level + 1)
    for f in files[:2]:                                     # Show only first 2 files per folder
        print(f"{sub_indent}{f}")
```

Lets count the number of pictures per folder and look at one sample image from each subfolder:

```python
# Function to load a sample image from each subfolder
def load_sample_images_from_folders():
    subfolders = ['not fractured', 'test', 'train', 'val']
    sample_images = []
    
    for subfolder in subfolders:
        img_path = os.path.join(dataset_path, subfolder, 'img')
        if os.path.exists(img_path):
            image_files = [f for f in os.listdir(img_path) if f.endswith('.jpg')]
            print(f"Found {len(image_files)} images in {subfolder}")
            if image_files:
                sample_images.append((os.path.join(img_path, image_files[0]), subfolder))
        else:
            print(f"Path does not exist: {img_path}")
    
    if not sample_images:
        print("No images found.")
        return 

    # Create subplots if sample_images is not empty
    fig, axes = plt.subplots(1, len(sample_images), figsize=(15, 5))
    for ax, (img_path, subfolder) in zip(axes, sample_images):
        img = Image.open(img_path)
        ax.imshow(img, cmap='gray')
        ax.set_title(os.path.basename(img_path))
        ax.axis('off')
        ax.text(0.5, -0.1, subfolder, ha='center', va='center', transform=ax.transAxes, fontsize=12, color='black')

    plt.show()
```

```python
# Display sample images from each subfolder
load_sample_images_from_folders()
```

Lets look at the annotations (bounding boxes) of an example from the train set.

```python
# Function to load and display an example image with bounding box annotations
def show_image_with_and_without_annotations(image_filename):
    # Define the path for the image and the corresponding annotation (JSON)
    img_path = os.path.join(dataset_path, 'train', 'img', image_filename)
    ann_path = os.path.join(dataset_path, 'train', 'ann', image_filename + '.json')  # Append '.json' to the image name

    # Check if the image and annotation file exist
    if not os.path.exists(img_path):
        print(f"Image file {image_filename} not found.")
        return

    if not os.path.exists(ann_path):
        print(f"Annotation file for {image_filename} not found.")
        return

    # Open the image
    img = Image.open(img_path)

    # Create a copy of the image for annotation purposes
    img_with_annotations = img.copy()
    draw = ImageDraw.Draw(img_with_annotations)

    # Load the annotation data
    with open(ann_path, 'r') as ann_file:
        annotation_data = json.load(ann_file)

    # Loop through each object in the "objects" key and draw the bounding boxes
    for obj in annotation_data.get('objects', []):
        if obj['geometryType'] == 'rectangle':  # Handling rectangular bounding boxes
            points = obj['points']['exterior']
            x1, y1 = points[0]
            x2, y2 = points[1]
            draw.rectangle([x1, y1, x2, y2], outline="red", width=3)  # Red rectangle

        elif obj['geometryType'] == 'polygon':  # Handling polygon bounding boxes
            points = obj['points']['exterior']
            flat_points = [coord for point in points for coord in point]  # Flatten the list of points
            draw.polygon(flat_points, outline="blue", width=3)  # Blue polygon

    # Display the images side by side (without bounding boxes on the left and with bounding boxes on the right)
    fig, axes = plt.subplots(1, 2, figsize=(20, 10))
    
    # Left image (without bounding boxes)
    axes[0].imshow(img)
    axes[0].set_title(f"Without Bounding Boxes")
    axes[0].axis('off')

    # Right image (with bounding boxes)
    axes[1].imshow(img_with_annotations)
    axes[1].set_title(f"With Bounding Boxes")
    axes[1].axis('off')

    plt.show()
```

```python
# Display an example image with and without bounding boxes
show_image_with_and_without_annotations("IMG0002652.jpg")  
```
